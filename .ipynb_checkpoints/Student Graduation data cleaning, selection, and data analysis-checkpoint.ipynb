```python
#Univeriante Selection


#apply SelectKBest class to extract top 10 best features
best_features = SelectKBest(score_func = chi2, k=10)
fit = best_features.fit(x, y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)


#concat two dataframes for better visualization 
feature_Scores = pd.concat([dfcolumns,dfscores],axis=1)
feature_Scores.columns = ['Specs','Score']  #naming the dataframe columns
print(feature_Scores.nlargest(10,'Score'))  #print 10 best features
```

# Elbow Analysis

```python
Nc = range(1, 20)

kmeans = [KMeans(n_clusters=i) for i in Nc]

kmeans

score = [kmeans[i].fit(y).score(y) for i in range(len(kmeans))]

score

pl.plot(Nc,score)

pl.xlabel('Number of Clusters')

pl.ylabel('Score')

pl.title('Elbow Curve')

pl.show()
```

# Support Vector Machine

```python

x_svm = x
y_svm = y

x_train_svm, x_test_svm, y_train_svm, y_test_svm = train_test_split(x_svm, y_svm, test_size = 0.40, random_state=0)
```

```python
SVM_model = SVC()

parameters = [{'kernel': ['rbf'], 
               'gamma': [1e-3, 1e-4], 
               'C': [1, 10, 100, 1000]}, 
              {'kernel': ['linear'], 
               'C': [1, 10, 100, 1000]}]
SVM_model = GridSearchCV(SVC(), parameters, cv = 5)
SVM_model.fit(x_train_svm, y_train_svm)
```

```python
print('Best score for training data:', SVM_model.best_score_,"\n") 

# View the best parameters for the model found using grid search
print('Best C:',SVM_model.best_estimator_.C,"\n") 
print('Best Kernel:',SVM_model.best_estimator_.kernel,"\n")
print('Best Gamma:', SVM_model.best_estimator_.gamma,"\n")
```

```python
best_estimator = SVM_model.best_estimator_
y_pred_svm = best_estimator.predict(x_test_svm)
#y_pred_label = enc.inverse_transform(y_pred)
```

```python
# Making the Confusion Matrix
#print(pd.crosstab(Y_test_label, Y_pred_label, rownames=['Actual Activity'], colnames=['Predicted Activity']))
from yellowbrick.classifier import ClassificationReport
classes = ['Failed', 'Passed']

class_report = ClassificationReport(best_estimator, classes=classes, support=True)

class_report.fit(x_train_svm, y_train_svm)  # Fit the visualizer and the model
class_report.score(x_test_svm, y_test_svm)  # Evaluate the model on the test data
g = class_report.poof()             # Draw/show/poof the data


print(confusion_matrix(y_test_svm, y_pred_svm))
print("\n")




data = confusion_matrix(y_test_svm, y_pred_svm)
df_cm = pd.DataFrame(data, columns=np.unique(y_test_svm), index = np.unique(y_test_svm))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'
plt.figure(figsize = (10,7))
sbn.set(font_scale=1.4)      #for label size
sbn.heatmap(df_cm, cmap="Blues", annot=True,annot_kws={"size": 16})# font size
print("\n")



print("Training set score for SVM: %f" % best_estimator.score(x_train_svm , y_train_svm))
print("Testing  set score for SVM: %f" % best_estimator.score(x_test_svm  , y_test_svm ))


SVM_model.score
```

```python
x = x_old[['G1', 'G2']]
y = y_old



pca = PCA(n_components=1)
pca.fit(x)



x = pca.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state=42) 
```

```python
linear_reg = LinearRegression( normalize= True)
linear_reg.fit(x_train, y_train)

y_pred_linear = linear_reg.predict(x_test)
```

```python
plt.scatter(x_test, y_test,  color='black')
plt.plot(x_test, y_pred_linear, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
ax1 = sbn.distplot(y_train, hist=False, color="r", label="Actual Value")
sbn.distplot(y_pred_linear, hist=False, color="b", label="predicted value" , ax=ax1)
```

```python
mean_squared_erro = (np.sqrt(mean_squared_error(y_test, y_pred_linear)))

r2 = r2_score(y_test, y_pred_linear)


print("Mean squared error result: %f" % mean_squared_erro)
print("r2 score result: %f" % r2)
```
